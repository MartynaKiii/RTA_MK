## Część 1: Wczytanie dane

### SparkSession — brama do Sparka

```python
from pyspark.sql import SparkSession

spark = (
    SparkSession.builder
    .appName("Lab2-Transactions")
    .getOrCreate()
)
spark.sparkContext.setLogLevel("WARN")
print(f"Spark {spark.version} — gotowy")
```

### Wczytanie pliku JSON

```python
df = spark.read.json("cwiczenia/RTA/transactions_10k.jsonl")

print(f"Liczba rekordów: {df.count()}")
df.printSchema()
df.show(10, truncate=False)
```

### Konwersja kolumny timestamp

```python
from pyspark.sql.functions import to_timestamp, col

df = df.withColumn("timestamp", to_timestamp(col("timestamp"), "yyyy-MM-dd HH:mm:ss"))

df.printSchema()  # timestamp powinien być teraz 'timestamp (nullable = true)'
```

## Część 2: Podstawowe agregacje

### Zadanie 2.1 — Liczba transakcji i suma przychodów per sklep


```python
from pyspark.sql.functions import count, sum as _sum, avg, round as _round

store_summary = (
    df.groupBy("store")
    .agg(
        count("tx_id").alias("liczba_tx"),
        _round(_sum("amount"), 2).alias("suma_PLN"),
        _round(avg("amount"), 2).alias("srednia_PLN"),
    )
    .orderBy("store")
)
store_summary.show()
```

### Zadanie 2.2 - Statystyki per kategoria

```python
from pyspark.sql.functions import min as _min, max as _max, sum as _sum, round as _round


category_stats = (
    df.groupBy("category")
    .agg(
        _sum("amount").alias("suma_kwot"),
        _min("amount").alias("min_kwota"),
        _max("amount").alias("max_kwota")
    )
    .orderBy("category")
)

category_stats.show()
```

## Część 3: Okna tumbling

### Zadanie 3.1 — Liczba transakcji per godzina (tumbling 1h)

```python
from pyspark.sql.functions import window

hourly = (
    df.groupBy(window("timestamp", "1 hour"))    # okno 1-godzinne
    .agg(
        count("tx_id").alias("liczba_tx"),
        _round(_sum("amount"), 2).alias("suma_PLN"),
    )
    .orderBy("window")
)
hourly.show(truncate=False)
```

```python
(
    hourly
    .select(
        col("window.start").alias("od"),
        col("window.end").alias("do"),
        "liczba_tx",
        "suma_PLN",
    )
    .show(truncate=False)
)
```

### Zadanie 3.2 Okna 30-minutowe per sklep

```python
from pyspark.sql.functions import window, count, col


store_windows_30m = (
    df.groupBy(window("timestamp", "30 minutes"), "store")
    .agg(
        count("tx_id").alias("liczba_tx"),
        _round(_sum("amount"), 2).alias("suma_PLN")
    )
    .orderBy("window", "store")
)

store_windows_30m.select(
    col("window.start").alias("od"),
    col("window.end").alias("do"),
    "store",
    "liczba_tx",
    "suma_PLN"
).show(truncate=False)
```

### Zadanie 3.3 — W której godzinie sklep “Kraków” miał najwyższy przychód?

```python
from pyspark.sql.functions import desc


krakow_peak_hour = (
    df.filter(col("store") == "Kraków")
    .groupBy(window("timestamp", "1 hour"))
    .agg(_round(_sum("amount"), 2).alias("suma_PLN"))
    .orderBy(desc("suma_PLN"))
)

krakow_peak_hour.select(
    col("window.start").alias("od"),
    col("window.end").alias("do"),
    "suma_PLN"
).show(1, truncate=False) 
```

## Część 4: Okna sliding

### Zadanie 4.1 — Okno 1h, krok 30 minut

```python
sliding = (
    df.groupBy(window("timestamp", "1 hour", "30 minutes"))  # szerokość 1h, krok 30min
    .agg(
        count("tx_id").alias("liczba_tx"),
        _round(_sum("amount"), 2).alias("suma_PLN"),
    )
    .select(
        col("window.start").alias("od"),
        col("window.end").alias("do"),
        "liczba_tx",
        "suma_PLN",
    )
    .orderBy("od")
)
sliding.show(truncate=False)
```

### Zadanie 4.2 — Porównaj tumbling vs sliding

```python
tumbling_rows = (
    df.groupBy(window("timestamp", "1 hour"))
    .agg(count("tx_id"))
    .count()
)
sliding_rows = (
    df.groupBy(window("timestamp", "1 hour", "30 minutes"))
    .agg(count("tx_id"))
    .count()
)
print(f"Tumbling (1h):          {tumbling_rows} okien")
print(f"Sliding  (1h / 30min):  {sliding_rows} okien")

# Odpowiedz w komentarzu: dlaczego sliding ma więcej wierszy?
# ODPOWIEDŹ: Sliding ma więcej wierszy, z tego względu, że  okna te rozpoczynaja się co 30 minut i jednocześnie nakladają się na siebie.
```

#### 2. Jaka jest różnica między groupBy("store") a groupBy(window(...), "store")?

##### Odpowiedź: groupBy("store") agreguje wszystkie dane z całego zbioru (np. z całego dnia) w jeden rekord dla każdego sklepu. groupBy(window(...), "store") dzieli dane na przedziały czasowe i liczy statystyki osobno dla każdego okna (np. co godzinę), pozwalając obserwować zmiany w czasie.

#### 3. W oknie sliding 1h/30min — ile okien zawiera transakcje z godziny 09:30?

## Praca domowa

#### 1. Znajdź godzinę, w której sklep Gdańsk miał najniższą średnią kwotę transakcji.

```python
from pyspark.sql.functions import avg

gdansk_min_avg = (
    df.filter(col("store") == "Gdańsk")
    .groupBy(window("timestamp", "1 hour"))
    .agg(_round(avg("amount"), 2).alias("srednia_kwota"))
    .orderBy("srednia_kwota") #domyslnie ASC sortowanie, wiec najnizsza bedzie na gorze
)
gdansk_min_avg.show(1)
```

#### 2. Policz ile transakcji per kategoria było w oknie 09:00–09:30.

```python

(
    df.groupBy(window("timestamp", "30 minutes"), "category")
    .agg(count("tx_id").alias("liczba_tx"))
    .filter("window.start = '2026-04-12 09:00:00'")
    .orderBy("category")
    .show(truncate=False)
)
```

#### 3. Zrób okno 15-minutowe i sprawdź w której ćwierćgodzinie był szczyt transakcji (łącznie dla wszystkich sklepów).

```python

(
    df.groupBy(window("timestamp", "15 minutes"))
    .agg(
        count("tx_id").alias("liczba_tx"),
        _round(_sum("amount"), 2).alias("suma_PLN"),
    )
    .orderBy("liczba_tx", ascending = False)
    .show(truncate=False)
)
```
